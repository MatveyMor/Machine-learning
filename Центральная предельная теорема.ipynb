# Равномерное распределение

Будем работать на отрезке [0,1]

a = 0

b = 1

Мат. ожидание для равномерного распределения равно $E[X] = (a+b)/2 = 0.5$

Дисперсия для такого распределения равна $D[X]=(b-a)^2/12 = 1/12$

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 1)

mean, var, skew, kurt = uniform.stats(moments='mvsk')

x = np.linspace(uniform.ppf(0.01),uniform.ppf(0.99), 100) #задаем массив из 100 элементов от 0 до 1
ax.plot(x, uniform.pdf(x),'r-', lw=5, alpha=0.6, label='pdf') #строим теоретическую плотность распределения

vals = uniform.ppf([0.001, 0.5, 0.999])
np.allclose([0.001, 0.5, 0.999], uniform.cdf(vals))
r = uniform.rvs(size=1000) #генерируем случайную выборку из 1000 значений с равномерным распределением

ax.hist(r, density=True, histtype='stepfilled', alpha=0.2) #строим гистограмму
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
n=5
sample5=np.array([]) #обозначим массив выборочных средних
i=1
while i <= 1000: #цикл генерации случайных выборок количества n из функции
    sample5i = np.random.uniform (0,1,n) #генерация выборки
    mean5i=np.mean(sample5i) #определение выборочного среднего (в.с.)
    sample5=np.append(sample5,mean5i) #добавляем в массив новое значение в.с.
    i+=1
#определим параметры нормального распределения описывающие массив 1000 в.с. при n=5
E=0.5
D=(1./12) #дисперсия равномерного распределения
D5=D/n #дисперсия для нормального распределения приближающего выборку средних
sigma5=math.sqrt(D5)#сигма для нормального распределения выборки средних
norm5 = sts.norm(E, sigma5)#определяем нормальное распределение с расчетными характеристиками равномерного
x = np.linspace(0,1,1000)
pdf5 = norm5.pdf(x[:1000])#получаем плотность распределения
#строим гистограмму массива средних при n=5 и плотность описывающую этот массив нормального распределения
fig, ax = plt.subplots(figsize=[12, 8])
plt.plot(x[:1000], pdf5, label='PDF(n=5)')
plt.hist(sample5, bins =80, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
plt.show()
```
