```python
import scipy as sp
import numpy as np

def f(x):
    return (np.sin(x/5.)*np.exp(x/10.) + 5*np.exp (-x/2.))

print (round(f(1),3), round(f(15),3))

A = np.array([[1., 1.], [1., 15.]])
b = np.array([f(1), f(15)])

h=np.linalg.solve(A, b)

%matplotlib inline
from matplotlib import pylab as plt

x = np.arange (-10, 10, 0.1)
y = np.sin(x/5.)*np.exp(x/10.) + 5*np.exp (-x/2.)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-10,10,0.1)
y = h[0] + h[1]*x
plt.plot(x,y)
plt.show()
```

```python
A = np.array([[1., 1., 1.], [1., 8., 64.],[1., 15., 225.]])
b = np.array([f(1), f(8), f(15)])
h=np.linalg.solve(A, b)

x = np.arange (-10, 10, 0.1)
y = h[0] + h[1]*x + h[2]*(x**2)
plt.plot(x,y)
plt.show()
```

```python
A = np.array([[1., 1., 1., 1.], [1., 4., 16., 64],[1., 10., 100., 1000.,],[1., 15., 225., 3375.]])
b = np.array([f(1), f(4), f(10), f(15)])
h=np.linalg.solve(A, b)
x = np.arange (-10, 10, 0.1)
y = h[0] + h[1]*x + h[2]*(x**2) + h[3]*(x**3)
plt.plot(x,y)
plt.show()
```
